Every great love starts with a great story...

```python
import math
%matplotlib inline
import matplotlib.pyplot as plt

def radius(theta):
    return (
        2 - 2 * math.sin(theta) 
        + math.sin(theta) * math.sqrt(abs(math.cos(theta))) / (math.sin(theta) + 1.4)
    )
theta = [t * math.pi / 180 for t in range(360)]
radius = [radius(t) for t in theta]
x = [r * math.cos(t) for r, t in zip(radius, theta)]
y = [r * math.sin(t) for r, t in zip(radius, theta)]

plt.plot(x, y)
```
